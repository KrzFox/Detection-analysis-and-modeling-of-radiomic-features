```python
import matplotlib.pyplot as plt
import pydicom
```

```python
filePath = "../../Env_3_7/data/square influence/image.dcm"
maskPath1 = "../../Env_3_7/data/masks short/45.dcm"
maskPath2 = "../../Env_3_7/data/masks short/99.dcm"
resultsPath = "../../Env_3_7/data/square influence/results"
```

```python
file = pydicom.dcmread(filePath)
mask1 = pydicom.dcmread(maskPath1)
mask2 = pydicom.dcmread(maskPath2)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 10))

ax[0].imshow(file.pixel_array*0.3 + mask1.pixel_array*80, cmap='gray')
ax[0].axis('off')

ax[1].imshow(file.pixel_array*0.3 + mask2.pixel_array*80, cmap='gray')
ax[1].axis('off')

plt.savefig(resultsPath + "/image")
plt.show()
```
